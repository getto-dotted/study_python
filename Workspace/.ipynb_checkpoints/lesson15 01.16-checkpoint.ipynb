```python
f = open('gender.csv')
data = csv.reader(f)

row = next(data)

M = []
F =[]

for i, d in enumerate(row):
    if '남' in d:
        M.append(i)
    elif '여' in d:
        F.append(i)
# print(min(M), max(M))
# print(min(F), max(F))
M_start = min(M) + 2
M_end = max(M) 
F_start = min(F) + 2
F_end = max(F) 

Mrs = []
Frs = []
for row in data:
    if '신도림' in row[0]:
        for i in row[M_start:M_end]:
            Mrs.append(int(i))
        for i in row[F_start:F_end]:
            Frs.append(int(i))
        break
        
# print(Mrs)
# print(Frs) 

Nrs = []

for i in Mrs:
    Nrs.append(i*(-1))
        
plt.barh(range(len(Nrs)), Nrs, label='남성')
plt.barh(range(len(Frs)), Frs, label='여성')

plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.legend()
plt.show()
```

```python
#지역 남녀 인구 총계 bar chart

f = open('gender.csv')
data = csv.reader(f)

row = next(data)

M = []
F =[]

for i, d in enumerate(row):
    if '남' in d:
        M.append(i)
    elif '여' in d:
        F.append(i)
# print(min(M), max(M))
# print(min(F), max(F))
M_start = min(M) + 2
M_end = max(M) 
F_start = min(F) + 2
F_end = max(F) 

Mrs = []
Frs = []
for row in data:
    if '신도림' in row[0]:
        for i in row[M_start:M_end]:
            Mrs.append(int(i))
        for i in row[F_start:F_end]:
            Frs.append(int(i))
        break
        
# print(Mrs)
# print(Frs) 

Msum = sum(Mrs)
Fsum = sum(Frs)
pop = [Msum, Fsum]

# plt.bar(range(len(pop)), pop)
color = ['crimson','darkcyan']

plt.pie(pop, labels=['남','여'], autopct='%.1f%%', colors = color, startangle=90)
plt.title('신도림 지역 남녀 비율')
plt.show()
```

```python
import csv
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
s_in = [0]*24
s_out = [0]*24
for row in data:
    row[4:] = map(int,row[4:])
    for i in range(24):
        s_in[i] += row[i*2+4]
        s_out[i] += row[i*2+5]


plt.figure(dpi = 300)
plt.rc('font', family ='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
