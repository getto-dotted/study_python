## 파일을 읽는 방법 4가지<br>
f = open('a.txt', 'r')<br>
text = f.read() -> 전체 읽어오기<br>
text = f.read(4) -> ()안의 용량만큼 읽어오기<br>
text = f.readline() -> 한줄 읽어오기<br>
text = f.readlines() -> 전체를 문장단위로 저장해서 읽어오기<br><br>
while line: -> 한줄씩 읽어와서 for문으로 전체를 쓰기<br>
&nbsp;&nbsp;&nbsp;&nbsp;line = f.readline()<br>
&nbsp;&nbsp;&nbsp;&nbsp;print(line)<br>
f.close()<br>

```python
#최고기온 구하기
f = open('seoul.csv', 'r', encoding='cp949')#파일오픈
data = csv.reader(f, delimiter=',')#csv파일 읽기
header = next(data)#컬럼부분을 읽어들여서 제외해준다/

max_temp = 0
max_date = ''
for row in data:  
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    
    if row[-1]> max_temp:
        max_temp = row[-1]
        max_date = row[0]
print(max_date, max_temp)   
f.close()  
```

```python
#최고기온을 히스토그램으로 표시
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.hist(result, bins=100)    
plt.show()
```

```python
#2월의 일 평균기온 데이터 히스토그램
#8월의 일 평균기온 히스토그램
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result2 = []
result8 = []
for row in data:
    if row[2] != '':
        if row[0].split('-')[1]=='02':
            result2.append(float(row[2]))
        elif row[0].split('-')[1]=='08':
            result8.append(float(row[2]))
        
#         if(row[2] != '') & (row[0].split('-')[1]=='02'):
#             result2.append(float(row[2]))
#         if(row[2] != '') & (row[0].split('-')[1]=='08'):
#             result8.append(float(row[2]))

plt.hist(result2, bins=100)    
plt.hist(result8, bins=100)    
plt.show()
```

```python
def cntBirth():
    ret = [] #결과를 저장할 리스트
    for y in range(1880, 2016): # 1880년 ~ 2015년
        cnt =0 # 신생아 인구를 저장할 변수
        fname = './names/yob{}.txt'.format(y) # 읽어올 파일이름 형식
        with open(fname, 'r') as f: 
            data = f.readlines() # 파일에서 데이터를 한줄씩 가져온다.
            for d in data:
                if d[-1] == '\n': # 한줄씩 가져온 데이터에서 마지막 줄바꿈 기호를 삭제한다.
                    d = d[:-1]
                    
                birth = d.split(',')[2]
                cnt += int(birth)
        ret.append((y,cnt))
    return ret
```

```python
#연도별 총 여아, 총남아 숫자를 구하기
def cntBirth():
    ret = [] #결과를 저장할 리스트
    
    for y in range(1880, 2016): # 1880년 ~ 2015년
        cntF =0 # 총여아 인구를 저장할 변수
        cntM =0 # 총남아 인구를 저장할 변수
        cnt =0 #총 인구를 저장할 변수
        fname = './names/yob{}.txt'.format(y) # 읽어올 파일이름 형식
        with open(fname, 'r') as f: 
            data = f.readlines() # 파일에서 데이터를 한줄씩 가져온다.
            for d in data:
                if d[-1] == '\n': # 한줄씩 가져온 데이터에서 마지막 줄바꿈 기호를 삭제한다.
                    d = d[:-1]
                    
                birth = d.split(',')[2]
                gender = d.split(',')[1]
                
                cnt += int(birth)
                
                if gender == 'M':
                    cntM += int(birth)
                else:
                    cntF += int(birth)    
        ret.append((y, cntF, cntM, cnt))        
        
        with open('bbyT.txt', 'w') as f:
            for year, cntF, cntM, cnt in ret:
                data = '{}년도 여아:{}명, 남아:{}명, 총합{}명\n'.format(year,cntF, cntM, cnt)           
                f.write(data) 
cntBirth()                
```

```python
#연도를 입력받은 후 그 연도의 남아, 여아 각각 인원이 가장 많은 이름 상위 10개씩 출력
def findN():
    ret = [] #결과를 저장할 리스트
    Fname = []
    Mname = []
    year = int(input('찾을 연도 입력'))
    
    fname = './names/yob{}.txt'.format(year)
    with open(fname, 'r') as f:
        data = f.readlines()
        
        for d in data:
            if d[-1] == '\n': # 한줄씩 가져온 데이터에서 마지막 줄바꿈 기호를 삭제한다.
                d = d[:-1]
                
            
            pname = d.split(',')[0]
            gender = d.split(',')[1]    
            
            if len(Mname) < 10:
                if gender == 'M':
                    Mname.append(pname)
            if len(Fname) < 10:
                if gender == 'F':
                    Fname.append(pname)                               
        
        ret.append((year, Fname, Mname))        
        
        with open('Pname.txt', 'w') as f:
            for year, Fname, Mname in ret:
                data = '{}년도 여아이름 10위권:{}, 남아이름 순위권:{}\n'.format(year, Fname, Mname)           
                f.write(data)
findN()                
```

```python
#년월 매출합구하기

def cntIncome():
    ret = [] #결과를 저장할 리스트
    for m in range(1, 13): # 1~12월
        
        if m <10: # 달 수가 10월 이하일때 앞에 0을 붙여준다.
            m = '0{}'.format(m)        
        total =0 # 총수입을 저장할 변수
        fname = './csv_files/2017.{}.csv'.format(m) # 읽어올 파일이름 형식
        with open(fname, 'r', encoding='utf-8') as f:     
            header = f.readline() # 맨위 컬럼명 부분을 뺀다.
            data = f.readlines() # 파일에서 데이터를 한줄씩 가져온다.
            for d in data:                
                income = int(d.split(',')[1])
                
                total += income
                
        ret.append((m,total))        
        
    print(ret)
    
cntIncome()    
```

```python
#파일명과 단어를 받아서 찾기

def word_find(file, word):    
    f = open(file, 'r')

    data = f.read()
    data.lower()
    data = data.split()

    D = {}

    for c in data:
        if c not in D.keys():
            D[c] = 1
        else:
            D[c] +=1     
            
    return D.get(word)
word_find('./mydata.txt', 'of')
```
