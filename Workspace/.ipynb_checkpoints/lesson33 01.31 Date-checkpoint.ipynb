# .dt.함수()
날짜형식이라면 apply()없이도 전체적용 가능

```python
# banks['Closing Date'].apply(lambda x: x.quarter)
banks['Closing Date'].dt.year
```

```python
banks['closing_quarter'], banks['closing_year'] = (banks['Closing Date'].dt.quarter,banks['Closing Date'].dt.year)
```

```python
banks.head()
```

```python
# 연도별로 문을 닫은 은행의 갯수를 출력/차트
banks.groupby('closing_year').size()
```

```python
banks.groupby('closing_year').size().plot()
```

```python
banks.groupby('closing_quarter').size()
```

```python
banks.groupby('Closing Date').size().plot()
```
