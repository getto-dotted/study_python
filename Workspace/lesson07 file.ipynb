```python
def writeNread():
    text = input()
    
    f = open('FreeInput.txt', 'w')
    f.write(text)
    f.close()
    
    g=open('FreeInput.txt', 'r')
    gread = g.readlines()
    print(gread)
    g.close()

writeNread()
```

이미지, 영상 소리, 코드 => binary file => 'rb', 'wb'모드로 읽음

```python
# dddd.gif 파일을 비둘기_복사.jpg 복사저장

f = open('dddd.gif', 'rb')
g=open('비둘기_복사.jpg', 'wb')

pic = f.read()
g.write(pic)

f.close()
g.close()
```

```python
#파일의 용량이 너무 커서 조금씩 읽어야 할때

f = open('dddd.gif', 'rb')
g=open('비둘기_복사1.jpg', 'wb')

bufsize = 1024
pic = f.read(bufsize)

while pic:
    g.write(pic)
    pic = f.read(bufsize)
        

f.close()
g.close()
```
