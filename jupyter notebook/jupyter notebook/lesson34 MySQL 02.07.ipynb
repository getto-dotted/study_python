```python
df.groupby('month').agg({'amount':sum,'amount':np.mean,'amount':np.size})
```

# .agg() 1개의 컬럼에 다수의 함수를 적용할때는 
# 원하는 함수들을 리스트로 묶어야 한다.

```python
df.groupby('month').agg({'amount':[sum,np.mean,np.size]})
```

```python
sql = '''select month(payment_date), sum(amount) '전체매출', avg(amount) '평균매출', count(amount) '결제횟수' 
from payment group by month(payment_date);'''
pd.read_sql(sql, db)
```

```python
# 렌탈횟수가 가장 많은 고객 id, 총매출액
sql = '''select customer_id, count(*) as '렌탈횟수', sum(amount) '매출액'
from payment group by customer_id order by 렌탈횟수 desc limit 1;'''
pd.read_sql(sql, db)
```

```python
sql = '''select title, sum(amount) s
from payment p, film f,inventory i, rental r 
where f.film_id=i.film_id and r.inventory_id =i.inventory_id and r.rental_id=p.rental_id group by title order by s desc limit 1;'''
pd.read_sql(sql, db)
```

```python
# film테이블에서 2006년이나 2007년에 출시되었으면서 PG등급이거나 G등급인 영화제목
fdf[((fdf['rating']=='PG') | (fdf['rating']=='G'))&((fdf['release_year']==2006) | (fdf['release_year']==2007))]['title']
```

```python
fdf.head(1)
```

```python
fdf[(fdf['rating'].isin(['PG','G']))&(fdf['release_year'].isin(['2006','2007']))][['title','rating','release_year']]
```

```python
sql = '''select title
from film where rating in ('pg','g') and release_year in (2006,2007);'''
pd.read_sql(sql, db)
```

```python
# film 테이블에서 설명에 boring이 포함되면서 렌탈비용이 0.99인 영화제목, 설명, 등급
```
